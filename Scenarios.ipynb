```python
import numpy as np
import cvxpy as cvx
```

|        | Asset 1 | Asset 2| Asset 3 | Asset 4 | Req
| :----- | :------:| :----: | :-----: | :-----: | :--:
| Cost        | 2.0 | 3.0 | 1.0 | 0.5
| Scenario 1  | 0.2 | 1.0 | 0.1 | 0.5 | 10
| Scenario 2  | 0.5 | 1.2 | 1.0 | 0.8 | 20
| Scenario 3  | 1.0 | 0.2 | 1.3 | 1.2 | 15

```python
S = np.array(
    [
    [0.2, 1.0, 0.1, 0.5],
    [0.5, 1.2, 1.0, 0.8],
    [1.0, 0.2, 1.3, 1.2]
    ]
)
c = np.array([2,3,1,5])
r = np.array([10,20,15])
[m,d] = np.shape(S)
```

```python
x = cvx.Variable(d)
```

```python
obj = cvx.Minimize(c@x)
```

```python
constraints = [S@x >= r, x >=0]
```

```python
prob = cvx.Problem(obj,constraints)
```

```python
prob.solve()
```

```python
print(prob.value)
```

```python
print(np.round(x.value,2))
```
